# 18. Which Workclass Getting The Highest Salary?

```python
data.columns
```

```python
data[['workclass','income']]
```

```python
data
```

```python
# 'income' column contains string values like '<=50K' and '>50K', you need to convert them to numeric values first.
data['income'] = data['income'].map({'<=50K': 0, '>50K': 1})
print(data['income'].dtype)
```

```python
result = data.groupby('workclass')['income'].mean().sort_values(ascending=False)
result
```
